```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Downloads/correlation_btw_funds_raised_vs_percent_lay_off.csv")
```

```python
df['funds_raised_clean'] = (
    df['funds_raised_in_millions']
    .str.replace('[$,]', '', regex=True)        
    .replace('N/A', pd.NA)                         
    .astype(float)                                 
)
```

```python
correlation = df['funds_raised_clean'].corr(df['percentage_laid_off'])
print("Correlation:", correlation)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='funds_raised_clean',
    y='percentage_laid_off'
)
plt.title('Funds Raised vs. Percentage Laid Off')
plt.xlabel('Funds Raised (in Millions)')
plt.ylabel('Percentage Laid Off')
plt.show()
```
